```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
files_csv = "../project_1/Resources/data_cleaned_2021.csv"
```

```python
df = pd.read_csv(files_csv)
df.head()
```

```python
data_df = df.drop(columns=['Competitors', 'Founded'])
data_df.head()
```

```python
data_df[['Min Salary', 'Max Salary']] = data_df['Salary'].str.split('-', expand=True)
data_df.head()

```

```python
data_df['Min Salary'] = data_df['Min Salary'].str.replace('$', '')
data_df['Min Salary'] = data_df['Min Salary'].str.replace('K', '')
data_df['Man Salary'] = data_df['Max Salary'].str.replace('$','')
data_df['Man Salary'] = data_df['Max Salary'].str.replace('K','')
data_df
```
